```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#pose= recorded velocity, position, and orientation. Sampled at 50Hz.
pose = r'C:\Users\srida\nuscenes2mcap\data\can_bus\scene-0103_pose.json'

#vehicle_moniter = recorded speed, throttle, and braking. Sampled at 2Hz(=40 data points)
vehicle_moniter= r'C:\Users\srida\nuscenes2mcap\data\can_bus\scene-0103_vehicle_monitor.json'


#take both data, resample and concatenate in single DataFrame

df1= pd.read_json(pose) #to read json in pandas
df2= pd.read_json(vehicle_moniter) #to read json in pandas
```

```python
#Json Pose - from list [] to seperate xyz column

df1[['accel_x', 'accel_y', 'accel_z']] = pd.DataFrame(df1['accel'].tolist(), index=df1.index)
df1[['vel_x', 'vel_y', 'vel_z']] = pd.DataFrame(df1['vel'].tolist(), index=df1.index)
df1[['pos_x', 'pos_y', 'pos_z']] = pd.DataFrame(df1['pos'].tolist(), index=df1.index)

#drop the original columns
df1.drop(columns=['accel', 'vel', 'pos'], inplace=True)
```

```python
# Convert 'utime' from microseconds(us) to datetime
df1['timestamp'] = pd.to_datetime(df1['utime'], unit='us')
df1.set_index('timestamp', inplace=True) #set timestamp as index# inplace =true(change of state or modify original data)
df1.drop(columns=['utime'], inplace=True) #drop utime column


# Convert 'utime' from microseconds(us) to datetime
df2['timestamp'] = pd.to_datetime(df2['utime'], unit='us')
df2.set_index('timestamp', inplace=True) #set timestamp as index# inplace =true(change of state or modify original data)
df2.drop(columns=['utime'], inplace=True) #drop utime column
```
